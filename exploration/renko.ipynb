```python
import pandas as pd
import talib
import numpy as np

# Generate 20 days of sample data
np.random.seed(0)
dates = pd.date_range(start='2023-01-01', periods=20)
high = np.random.uniform(low=100, high=200, size=20)
low = high - np.random.uniform(low=5, high=15, size=20)
close = low + np.random.uniform(low=1, high=10, size=20)

data = {
    'Date': dates,
    'High': high,
    'Low': low,
    'Close': close
}
df = pd.DataFrame(data)

df['ATR'] = talib.ATR(df['High'], df['Low'], df['Close'], timeperiod=2)
print(df)
```
