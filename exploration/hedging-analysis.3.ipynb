```python
def trade_returns(file):
    with open(PATH + file, 'rb') as f:
        df = pkl.load(f)
    df = df['results'].copy()
    trades = df[df.signal!=0].copy()
    trades_in_streak = trades.groupby('streak_no').apply(lambda x: x.iloc[[-1]])['trade_no'].reset_index().drop(columns='level_1')
    trades_in_streak.rename(columns=dict(trade_no='trades_in_streak'), inplace=True)

    trades = pd.merge(left=trades, right=trades_in_streak, on='streak_no')
    trades['return'] = trades['ac_bal'].shift(-1) - trades['ac_bal']
    trades.trades_in_streak = np.where((trades.trade_no == 5) & (trades['return'] < 0), 6, trades.trades_in_streak)
    trades.trades_in_streak = np.where((trades.trade_no != 5) & (trades.trades_in_streak == 5), np.nan, trades.trades_in_streak)
    trades.trades_in_streak = trades.trades_in_streak.bfill()

    return trades
```

```python
def sharpe_ratio(returns: pd.Series, risk_free_rate: float=0) -> float:
    return (returns.mean() - risk_free_rate) / returns.std()

def sortino_ratio(returns: pd.Series, risk_free_rate: float=0, target_return: float=0) -> float:
    downside_returns = np.array([ret - risk_free_rate for ret in returns if ret < target_return])
    std_dev_downside = np.std(downside_returns, ddof=1) if len(downside_returns) > 0 else 0
    return (np.mean(returns) - risk_free_rate) / std_dev_downside if std_dev_downside != 0 else np.nan

def max_drawdown(returns: pd.Series) -> float:
    '''
    cum_returns: Calculates the cumulative returns using np.cumprod(1 + returns).
    peaks: Finds the peaks in the cumulative returns using np.maximum.accumulate(cum_returns).
    drawdowns: Calculates the drawdowns for each point by subtracting the peaks from the cumulative returns and dividing by the peaks.
    max_drawdown: Identifies the maximum drawdown by finding the minimum drawdown value using np.min(drawdowns).'''
    cum_returns = np.cumprod(1 + returns.values)
    peaks = np.maximum.accumulate(cum_returns)
    drawdowns = (cum_returns - peaks) / peaks
    max_drawdown = np.min(drawdowns)
    return max_drawdown

def calmar_ratio(returns: pd.Series):
    max_drawdown_value = max_drawdown(returns)
    annual_return = np.mean(returns) * 311  # Assuming 311 trading days per year for forex. Its usually 252 for stocks
    return annual_return / abs(max_drawdown_value)

def cagr(ac_bal: pd.Series, years: float) -> float:
    return (ac_bal.iloc[-1] / ac_bal.iloc[0]) ** (1 / years) - 1
```

```python
def show_results(data: pd.DataFrame):
    df = data.copy()
    results = list()

    results.append(['return_%', (df['ac_bal'].iloc[-1] - df['ac_bal'].iloc[0]) / df['ac_bal'].iloc[0]])
    results.append(['final_ac_bal', df['ac_bal'].iloc[-1]])
    results.append(['cagr', cagr(df['ac_bal'], 8)])
    results.append(['max_ac_bal', df['ac_bal'].max()])
    results.append(['min_ac_bal', df['ac_bal'].min()])
    results.append(['max_margin_used', df['margin_used'].max()])
    results.append(['total_streaks', df['streak_no'].iloc[-1]])
        
    # trades = df[df.signal!=0].copy()
    streaks = df[(df.signal!=0) & (df.trade_no == 1)].copy()
    streaks['return'] = streaks['ac_bal'].shift(-1) - streaks['ac_bal']
    streaks['return_%'] = streaks['ac_bal'].pct_change().shift(-1)
    streaks['index'] = streaks.index
    streaks['duration'] = streaks['index'].shift(-1) - streaks['index']
    streaks.dropna(inplace=True)
    # print(trades[['ac_bal', 'return', 'return_%']].head())

    results.append(['sharpe_ratio', sharpe_ratio(streaks['return_%'])])
    results.append(['sortino_ratio', sortino_ratio(streaks['return_%'])])
    results.append(['calmar_ratio', calmar_ratio(streaks['return_%'])])
    results.append(['max_drawdown', max_drawdown(streaks['return_%'])])

    results.append(['total_streaks', streaks.shape[0]])
    results.append(['avg_streak', streaks['return'].mean()])
    
    
    wins = streaks[streaks['return'] > 0]
    results.append(['total_wins', wins.shape[0]])
    results.append(['max_win', wins['return'].max()])
    results.append(['avg_win', wins['return'].mean()])
    results.append(['avg_win_duration', wins['duration'].mean()])
    results.append(['win_%', wins.shape[0] / streaks.shape[0]])

    losses = streaks[streaks['return'] < 0]
    results.append(['total_losses', losses.shape[0]])
    results.append(['max_loss', losses['return'].min()])
    results.append(['avg_loss', losses['return'].mean()])
    results.append(['avg_loss_duration', losses['duration'].mean()])
    results.append(['loss_%', losses.shape[0] / streaks.shape[0]])
    
    longs = streaks[streaks.signal == 1]
    results.append(['total_long', longs.shape[0]])
    results.append(['avg_long', longs['return'].mean()])
    results.append(['avg_long_duration', longs['duration'].mean()])

    long_wins = longs[longs['return'] > 0]
    results.append(['total_long_wins', long_wins.shape[0]])
    results.append(['avg_long_win', long_wins['return'].mean()])
    results.append(['avg_long_win_duration', long_wins['duration'].mean()])

    long_losses = longs[longs['return'] < 0]
    results.append(['total_long_losses', long_losses.shape[0]])
    results.append(['avg_long_loss', long_losses['return'].mean()])
    results.append(['avg_long_loss_duration', long_losses['duration'].mean()])

    shorts = streaks[streaks.signal == -1]
    results.append(['total_short', shorts.shape[0]])
    results.append(['avg_short', shorts['return'].mean()])
    results.append(['avg_short_duration', shorts['duration'].mean()])

    short_wins = shorts[shorts['return'] > 0]
    results.append(['total_short_wins', short_wins.shape[0]])
    results.append(['avg_short_win', short_wins['return'].mean()])
    results.append(['avg_short_win_duration', short_wins['duration'].mean()])

    short_losses = shorts[shorts['return'] < 0]
    results.append(['total_short_losses', short_losses.shape[0]])
    results.append(['avg_short_loss', short_losses['return'].mean()])
    results.append(['avg_short_loss_duration', short_losses['duration'].mean()])

    results.append(['expectancy', wins.shape[0] / streaks.shape[0] * wins['return'].mean() + \
                    losses.shape[0] / streaks.shape[0] * losses['return'].mean()])

    print(tabulate(results))
```
